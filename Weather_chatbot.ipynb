```python
import tkinter as tk
from tkinter import messagebox
import requests

# Function to get weather data
def get_weather(city):
    API_KEY = 'ce15601d4acc48fba9c71617241010'  # Replace with your actual API key
    BASE_URL = f"http://api.weatherapi.com/v1/current.json?key={API_KEY}&q={city}&aqi=no"
    
    response = requests.get(BASE_URL)
    
    if response.status_code == 200:
        data = response.json()
        weather_info = {
            "location": data['location']['name'],
            "region": data['location']['region'],
            "country": data['location']['country'],
            "temperature_c": data['current']['temp_c'],
            "condition": data['current']['condition']['text'],
            "wind_kph": data['current']['wind_kph'],
            "humidity": data['current']['humidity']
        }
        return weather_info
    else:
        return {"error": "Sorry, I couldn't find that location."}

# Function to give weather advice
def get_weather_advice(condition, temperature):
    condition = condition.lower()
    
    if "rain" in condition:
        return "Don't forget to bring an umbrella!"
    elif "sunny" in condition and temperature < 30:
        return "It's a nice day!"
    elif "wind" in condition:
        return "Don't forget your jacket!"
    elif "clear" in condition:
        return "Its clear. Go outside!"
    elif temperature > 35:
        return "Too hot! Stay indoors!!"
    elif temperature <= 5:
        return "Freezing outside! Wear layers"
    else:
        return ""

# Function to handle the button click event
def get_weather_info():
    city = city_entry.get()  # Get the city name from the text entry
    weather_data = get_weather(city)
    
    if "error" not in weather_data:
        advice = get_weather_advice(weather_data['condition'], weather_data['temperature_c'])
        result_label.config(text=(
            f"Location: {weather_data['location']}, {weather_data['region']}, {weather_data['country']}\n"
            f"Temperature: {weather_data['temperature_c']}Â°C\n"
            f"Condition: {weather_data['condition']}\n"
            f"Wind Speed: {weather_data['wind_kph']} kph\n"
            f"Humidity: {weather_data['humidity']}%\n"
            f"{advice}"
        ))
    else:
        messagebox.showerror("Error", weather_data['error'])

# Create the main window
root = tk.Tk()
root.title("Weather Chatbot")

# Set the size of the window
root.geometry("400x300")

# Add an entry widget to input the city name
city_label = tk.Label(root, text="Enter City:")
city_label.pack(pady=10)

city_entry = tk.Entry(root, width=40)
city_entry.pack(pady=5)

# Add a button to get the weather
weather_button = tk.Button(root, text="Get Weather", command=get_weather_info)
weather_button.pack(pady=10)


result_label = tk.Label(root, text="", justify="left", font=("Arial", 12))
result_label.pack(pady=20)

# Run the tkinter event loop
root.mainloop()
```
